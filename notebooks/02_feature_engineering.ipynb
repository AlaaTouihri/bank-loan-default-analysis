```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:\\Users\\\hp\\Desktop\\bank-loan-default-analysis\\Data\\raw\\credit_card_default.csv")
```

```python
TARGET_COL = "default.payment.next.month"
```

```python
# Colonnes PAY_0 Ã  PAY_6 : historique des retards de paiement
pay_cols = ["PAY_0","PAY_2","PAY_3","PAY_4","PAY_5","PAY_6"]

df["avg_payment_delay"] = df[pay_cols].replace({-1:0}).mean(axis=1)
```

```python
df["months_in_delay"] = (df[pay_cols] > 0).sum(axis=1)
```

```python
bill_cols = ["BILL_AMT1","BILL_AMT2","BILL_AMT3","BILL_AMT4","BILL_AMT5","BILL_AMT6"]

df["avg_bill_amount"] = df[bill_cols].mean(axis=1)
df["credit_utilization_ratio"] = df["avg_bill_amount"] / df["LIMIT_BAL"]
```

```python
pay_amt_cols = ["PAY_AMT1","PAY_AMT2","PAY_AMT3","PAY_AMT4","PAY_AMT5","PAY_AMT6"]

df["total_paid_amount"] = df[pay_amt_cols].sum(axis=1)
df["avg_payment_ratio"] = df["total_paid_amount"] / df["avg_bill_amount"]
```

```python
df["payment_consistency_score"] = df[pay_amt_cols].std(axis=1)
```

```python
df[["avg_payment_delay", "months_in_delay", "credit_utilization_ratio",
    "total_paid_amount", "avg_payment_ratio", "payment_consistency_score"]].head()
```

## Feature Engineering Summary

- **avg_payment_delay** : average delay in months (risk indicator)
- **months_in_delay** : number of months with delayed payment
- **credit_utilization_ratio** : average bill / credit limit
- **total_paid_amount** : total amount paid over 6 months
- **avg_payment_ratio** : proportion of bill actually paid
- **payment_consistency_score** : standard deviation of monthly payments (stability)

These features are designed to capture customer behavior and financial stability, improving model performance and interpretability.
