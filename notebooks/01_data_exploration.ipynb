## Key Insights from Exploratory Data Analysis

- Payment delay variables are the strongest indicators of default risk.
- Clients with lower credit limits tend to default more frequently.
- Demographic variables have a moderate impact compared to behavioral variables.
- The dataset is imbalanced, which must be considered during modeling.
